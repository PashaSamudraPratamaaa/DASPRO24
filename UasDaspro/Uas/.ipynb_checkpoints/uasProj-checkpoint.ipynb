```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Baca file Netflix
try:
    df = pd.read_csv("netflix_titles.csv")
except FileNotFoundError:
    print("File not found")
```

```python
# Hapus baris dengan release_year kosong
df = df.dropna(subset=['release_year'])

# Ubah kolom release_year menjadi integer
df['release_year'] = df['release_year'].astype(int)

# Hitung jumlah rilisan per tahun
release_counts = df['release_year'].value_counts().sort_index()
```

```python
# Line plot tren jumlah rilisan
plt.figure(figsize=(12, 5))
plt.plot(release_counts.index, release_counts.values, marker='o', label='Jumlah Rilisan')
plt.title("Tren Jumlah Rilisan Netflix per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Rilisan")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Histogram distribusi jumlah rilisan
plt.figure(figsize=(8, 5))
plt.hist(release_counts.values, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Distribusi Jumlah Rilisan per Tahun")
plt.xlabel("Jumlah Rilisan")
plt.ylabel("Frekuensi")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Boxplot untuk outlier tahun rilisan
plt.figure(figsize=(10, 2))
plt.boxplot(release_counts.values, vert=False)
plt.title("Boxplot Jumlah Rilisan per Tahun")
plt.xlabel("Jumlah Rilisan")
plt.tight_layout()
plt.show()
```

```python
# Error bar: mean dan standard deviation
mean_releases = release_counts.mean()
std_releases = release_counts.std()
plt.errorbar(['Rilisan'], [mean_releases], yerr=[std_releases], fmt='o', capsize=5, label='Mean ± Std Dev')
plt.title('Mean dan Standard Deviation Jumlah Rilisan per Tahun')
plt.ylabel('Jumlah Rilisan')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Linear Regression: prediksi tren jumlah rilisan
X = release_counts.index.values.reshape(-1, 1)
y = release_counts.values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.figure(figsize=(12, 5))
plt.scatter(X.flatten(), y, label='Data Aktual', color='blue')
plt.plot(X.flatten(), y_pred, label='Garis Regresi', color='red', linestyle='--')
plt.title('Regresi Linear Jumlah Rilisan Netflix per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Rilisan')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
